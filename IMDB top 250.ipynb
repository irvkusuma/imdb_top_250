```python
import requests as req

#put whole site to variable
open_page = req.get('https://www.imdb.com/chart/top?sort=ir,desc&mode=simple&page=1a')

#close existing connection
open_page.close()
```

```python
from bs4 import BeautifulSoup as soup

#make code easy to read
page_html = soup(open_page.content,'html.parser')
```

```python
table = page_html.find_all('table',{'class':'chart full-width'})
```

```python
a = table[0]
```

```python
headCol = a.find_all('th')
column = [Z.get_text() for Z in headCol]
```

```python
movie_title = a.find_all('td',{'class':'titleColumn'})
```

```python
movie_rating = a.find_all('td',{'class':'ratingColumn imdbRating'})
```

```python
title = [Z.get_text() for Z in movie_title]
ratings = [Y.get_text() for Y in movie_rating]
```

```python
import pandas as pd
import csv
movieDB = pd.DataFrame({
    'Title': title,
    'Ratings': ratings
})

file = 'E:\SampleData\IMDB top 250.csv'


movieDB['Ratings'].str.strip()
movieDB['Ratings'].replace(r'\s+|\\n', ' ', regex=True, inplace=True) 
movieDB['Title'].replace(r'\s+|\\n', ' ', regex=True, inplace=True) 

new_column = movieDB['Title'].str.split(pat = '.', n = 1, expand = True)
movieDB['No.'] = new_column[0]
movieDB['Title'] = new_column[1]

movieDB['Title'].replace('.','', inplace = True)

fi = movieDB.to_csv(file,sep = ';', index = False, header = True)
```
